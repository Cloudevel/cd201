## El Modelo del objeto documento (DOM).

El DOM permite a los desarrolladores web tener acceso y poder manipular páginas HTML como si fueran documentos XML, a pesar de sus oigenes el DOM se ha vuelto una utilidad que puede ser aprovechada por otros lenguajes como Java y PHP.

 
### Árbol de nodos.

Comúnmente dentro de la programación web en JavaScript, se requiere tener acceso a las páginas para realizar diversas manipulaciones, por ejemplo valores almacenados por algunos elementos o creación de algunos elementos dinámicos.

Es posible manipular uno o varios elementos de un documento HTML mendiante el DOM. El DOM transforma todos los documentos HTML en un conjunto de elementos a los cuales les denomina "nodos", los cuales están interconectados y representan los contenidos de las páginas web, definiéndose así como "árbol de nodos".

La raíz del árbol de nodos siempre será la misma un nodo especial llamado "Documento", incluso es posible que se tomen a la ventana como nodo raíz, haciendo referencia al navegador.

A partir del nodo raíz, cada etiqueta HTML se transforma en un nodo de tipo "Elemento". La conversión de etiquetas a nodos se realizan de manera jerárquica de esta forma del nodo raíz derivan HEAD y BODY, a partir de esta derivación cada etiqueta se transforma en un nodo que deriva de su "etiqueta padre".

Durante la transformación de las etiquetas HTML, se generan dos nodos:

* Elemento: Corresponde a la propia etiqueta
* Texto: Contenido de la etiqueta

### Tipos de nodo.

HTML solo maneja los siguientes nodos:

    Document, nodo raíz del que derivan todos los nodos.
    Element, representa a cada una de las etiqeutas
    Attr, define el valor de cada etiqueta XHTML
    Text, contiene el texto que está dentro de una etiqueta
    Comment, comentarios incluidos en la página
 
### Acceso directo a los nodos

Una vez creado nuestro árbol de nodos DOM nos permite acceder de manera directa a cualquier nodo, lo cual es equivalente a acceder a un elemento de la página, donde es posible realizar tareas como:

    Acceder a valor de un elemento
    Establecer valores a un elemento
    Mover un elemento de la página
    Crear o eliminar elementos

La manera mas común que DOM nos proporciona para accesar a sus nodos, es partiendo del nodo raíz, descendiendo por sus ramas hasta llegar al nodo que deseamos, pasando por nodos padres e hijos, sin embargo, cuando se quiere llegar a un nodo específico se puede acceder a el a través de algunas funcionesm que veremos a continuación.

Cabe mencionar que las operaciones con nodos (creación, eliminación, modificación), solamente pueden ser realizados cuando la página se haya cargado por completo.
