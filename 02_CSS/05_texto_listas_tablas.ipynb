## Tipografía.


## Selección del tipo de fuente con *font-family*.

CSS permite seleccionar el tipo de letra que se utilizará en diversos elementos.

La propiedad *font-family*.

Permite elegir la tipografía a partir de las diversas familias de fuentes. 

Cabe señalar que es necesario que el navegador tenga instaladas las tipografías a las que se hace referencia.

## Importación de una fuente tipográfica externa.


En caso de utilizar una tipografía poco convencional que no se encuentre instalada en el equipo desde el cual el navegador accede al documento, es posible indicarle que puede descargarla e instalarla mediante *@import* y *@font-face*.

Los parámetros que utiliza @font-face para descargar e instalar la tipografía que se le indica es:

    font-famiiy, tal como se utiliza normalmente.
    src, que debe incluir la URL desde la cual se descargará el archivo de instalación de la tipografía.
    format es el formato en le que la tipografía es entregada y a saber son:
        eot
        woff
        ttf
        svg

## Definición del tamaño de una fuente.

Con este atributo es posible definir el tamaño de la tipografía de letra para los elementos seleccionados.

Como ya se mencionó con anterioridad, es posible utilizar varias métricas para definir el tamaño una fuente tipográfica.

    Pixeles.
    Porcentajes con respecto al tamaño de la tipografía que el navegador tiene configurada por defecto.
    EMS. Un em equivale al ancho de un caracter m correspondiente a a tipografía que el navegador tiene configurada por defecto.



Cabe hacer notar que algunos estos formatos tienen soporte limitado entre navegadores.

Ajustes de texto.

Caracteres en negrita.

font-weight con los valores:

    normal
    bold

Caracteres en itálica.

font-style con los valores:

    normal
    italic
    oblique

Transformación de texto:

text-transform con los valores:

    uppercase convierrte el texto a mayúsculas.
    lowecase convierrte el texto a minúsculas.
    capitalize la primera letra de la palabra en mayúsculas.

Línea horizontal.

text-decoration con los valores:

    none
    underline subrrayado.
    overline raya encima del texto.
    line-through raya cruzando el texto.
    blink texto parpadeante

Caracterísitcas de párrafo.

line-height altura de la línea.

letter spacing espaciado entre letras.

word-spacing espaciado entre palabras.

text-align con los valores:

    left
    right
    center
    justify

vertical-align con los valores:

    baseline
    sub
    super
    top
    text-top
    middle
    text-botton

text-indent indentación del párrafo.
Selección de elementos en un párrafo

:first-letter primera letra del texto.

:first-line primera línea del texto.

 :link texto de hipervínculo.

:visited hipervínculo visitados.

:hover cuando el cursor pasa encima del texto.

:focus cuando el navegador detecta una posible interacción con el elemento.

Propiedades de las listas.

Listas no ordenadas.

Para definir el tipo de "balazo" o marcador se utiliza list-style-type con los siguientes valores:

    none
    disc
    circle 
    square

Listas ordenadas.

Para definir el tipo de numerador también se utiliza list-style-type con los siguientes valores:

    decimal
    decimal-leading-zero
    lower-alpha-
    upper-alpha
    lower-roman
    upper-roman

imágenes en los "balazos".

Es posible utilizar imágenes que sustituyan a los caracteres clásicos de los balazos mediante la propiedad list-style-image que tendría la siguiente sintaxis:

list-style-image: url("URL de la imagen");

Posición de los balazos con respecto al párrafo de texto.

Para definir si hay una indentación del párrafo con respecto al balazo de una lista se uitliza list-style-position con los siguientes valores:

    outside que indica que el texto de las líneas posteriores a la primera no puede ponerse debajo del balazo (se indentan). 
    inside que indica que el texto de las líneas posteriores a la primera comienzan debajo del balazo (no se indentan). 

Propiedades de las tablas.

Todos los elementos de las tablas siguen el modelo de caja y las propiedades de texto anteriormente discutidas. Sin embargo poseen las siguientes propiedades adicionales:
Mostrar bordes en celdas vacías.

Para definir si las celdas vacías presentan bordes se utiliza la propiedad empty-cells con los siguientes valores:

    show que permite que se vean los bordes.
    hide que esconde los bordes.
    inherit que hereda el comportamiento si se trata de una tabla dentro de otra.

Espaciamiento entre los bordes de una tabla.

Por defecto siempre hay un pequeño espacio que separa los bordes de una celda con respecto a otra y para definir el espacioamento o si debe haberlo se utilizan border-spacing y broder-colapse.

border-spacing define el espaciado de los bordes.

border-colapse define si existe o no la separación entre los bordes mediante los siguientes valores.

    collapse no hay espacios entre bordes.
    separate sí hay espacios entre bordes.
